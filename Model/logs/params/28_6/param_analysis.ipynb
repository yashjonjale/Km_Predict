```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
import glob
import os

pattern = "/home/yashjonjale/Documents/intern_proj/Model/logs/model_5_res/best_params_0.8_*"

#find files with the pattern in the directory and sort them by creation time
files = glob.glob(pattern)
files.sort(key=os.path.getmtime)

```

```python
df = []
for file in files:
    with open(file) as f:
        data = json.load(f)
        df.append(data)
```

```python
df = pd.DataFrame(df)
```
